```python
# Creating the copy the dataset
hotel_newdata = hotel_data.copy()
```

<font size="4">**Managing Duplicate and Missing Values**</font>




```python
# checking for duplicate values
hotel_newdata[hotel_newdata.duplicated()].shape
```

```python
hotel_newdata.drop_duplicates(inplace = True)
```

```python
hotel_newdata.shape
```

```python
# checking the missing values 
hotel_newdata.isnull().sum()
```

```python
#Replace the null values by 0 in these columns
hotel_newdata['children'].fillna(0,inplace=True)
hotel_newdata['country'].fillna('na',inplace=True)
hotel_newdata['company'].fillna(0,inplace=True)
hotel_newdata['agent'].fillna(0,inplace=True)
```

```python
hotel_newdata.isnull().sum()
```

```python
#Creating new column 
hotel_newdata['total_guest'] = hotel_newdata['adults']+hotel_newdata['children']+hotel_newdata['babies']
```

```python
new_data = hotel_newdata[['lead_time','previous_cancellations','previous_bookings_not_canceled','booking_changes','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests','total_guest']]
```

<font size="6">**Exploratory Data Analysis**</font>




The correlation is between the numerical data.

```python
#Finding correlation 
corr_df = new_data.corr()
```

```python
corr_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(corr_df,annot = True,fmt='.2f', annot_kws={'size': 10},  vmax=.8, square=True)
```

From the above heatmap, the following deductions can be made 

1.   A little correlation exists between total people and adr, which is 0.39 as it makes sense that higher revenue equals higher adr.




<font size="6">**Q1- Which is the most booked hotel**</font>





```python
booked_hotel_df = hotel_newdata['hotel'].value_counts()
booked_hotel_df

booked_hotel_df.plot(kind = 'bar',color=['black', 'red',],figsize=(8,6))
plt.title('Booked Hotels')
plt.ylabel('Number of bookings')
plt.xlabel('Type of Hotel')
```

<font size="4">**Looking at the results above, we can see that city hotels have more booking than for resort hotels.**</font>





<font size="6">**Q2- which is busiest month in the year**</font>

```python
busiest_month_df = hotel_newdata['arrival_date_month'].value_counts()
busiest_month_df

busiest_month_df.plot(kind = 'bar',color=['black', 'orange',],figsize=(12,8))
plt.title('Bookings across hotels in a year')
plt.ylabel('Bookings')
plt.xlabel('Month')
```

<font size="4">**The following hotel occupancy rates are highest in August and July and lowest in November and January**<font>

<font size="6">**Q3- Which hotel has the most booking for each year?**</font>


```python
plt.subplots(figsize=(10,7))
sns.countplot(x='arrival_date_year', hue='hotel',  data=hotel_newdata).set_title('yearly bookings') ;
```

<font size="4">**From above in 2015 resort hotel has more booking than city hotel whereas,In 2016 city hotel has far more booking than resort hotel moreover,In 2017 city hotel still ahs more booking.**<font>

<font size="6">**Q4- Which country has the highest booking and guest?**</font>


```python
y= list(hotel_newdata.country.value_counts().head (10))
x= list(hotel_newdata.country.value_counts().head(10).index)
plt.plot(x, y,color = 'red')
plt.figure(figsize=[58,44])
plt.show()
```

<font size="4">**Above data shows PRT (Portugal) has the most guest, followed by GBR (United Kingdom), FRA (France), ESP (Spain) and DEU (Germany)**.<font>

<font size="6">**Q5- What is proportion of non-cancelled  and cancelled reservations.**</font>

> Indented block




```python
cancelled_d= hotel_newdata['is_canceled'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(cancelled_d, labels=['Not Cancelled', 'Cancelled'], autopct='%.1f%%', startangle=90)
```

<font size="4">**Above pie chart shows that there are 27.5% Cancelled and 72.5% Non Cancelled reservations**.<font>

<font size="6">**Q6- Which rooms in each hotel are the most frequently used?**</font>


```python
plt.figure(figsize=(12,7))
sns.countplot(data=hotel_newdata,x="reserved_room_type", hue="hotel")
plt.show()
```

<font size="4">**Above data shows that A room is most used in both hotels ,followed  by D room. However,C room in resort hotel has not used whereas,B room in city hotel is also not used.** <font>
